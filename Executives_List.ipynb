```python
import pandas as pd
import yaml

with open('executive_branch.yaml', 'r') as f:
    df_terms = pd.io.json.json_normalize(yaml.load(f), 'terms')

df_terms
```

```python
df_presidents = df_terms[df_terms.type == 'prez']
```

```python
year = 1789
year_parties_pres = []

while year < 2018:
    
    dem_count = 0
    rep_count = 0
    fed_count = 0
    whig_count = 0
    demrep_count = 0
    ind_count = 0
    oth_count = 0
    members_count = 0
    partyinpower = ""
    partyinpower_count = 0
    partyinpower_list = []
    
    for index, row in df_presidents.iterrows():
        party = row['party']
        year_start = row['start'][:5]
        if (year_start.find('-') > 0):
            year_start = year_start.split("-",1)[0]
        year_end = row['end'][:5]
        if (year_end.find('-') > 0):
            year_end = year_end.split("-",1)[0]
        r = range(int(year_start),int(year_end))
        if (int(year_start) <= year <= int(year_end)):
            if (party == "Democrat"):
                dem_count += 1
                partyinpower_list.append(party)
            elif (party == "Republican"):
                rep_count += 1
                partyinpower_list.append(party)
            elif (party == "Federalist"):
                fed_count += 1
                partyinpower_list.append(party)
            elif (party == "Whig"):
                whig_count += 1
                partyinpower_list.append(party)
            elif (party == "Democratic Republican"):
                demrep_count += 1
                partyinpower_list.append(party)
            elif (party == "Democratic-Republican"):
                demrep_count += 1
                partyinpower_list.append('Democratic Republican')
            elif (party == "Independent"):
                ind_count += 1
                partyinpower_list.append(party)
            else:
                oth_count += 1
                partyinpower_list.append(party)

    members_count = dem_count + rep_count + fed_count + whig_count + demrep_count + ind_count + oth_count
    
    year_count_list = [dem_count, rep_count, fed_count, whig_count, demrep_count, ind_count, oth_count]
    year_max = max(year_count_list)
    
    if (members_count >=2):
        partyinpower = partyinpower_list[1] + 's'
    elif (year_max == dem_count):
        partyinpower = 'Democrats'    
    elif (year_max == rep_count):
        partyinpower = 'Republicans'
        partyinpower_count = year_max
    elif (year_max == fed_count):
        partyinpower = 'Federalists'
        partyinpower_count = year_max
    elif (year_max == whig_count):
        partyinpower = 'Whigs'
        partyinpower_count = year_max
    elif (year_max == demrep_count):
        partyinpower = 'Democratic Republicans'
        partyinpower_count = year_max  
    elif (year_max == ind_count):
        partyinpower = 'Independents'
        partyinpower_count = year_max   
    elif (year_max == oth_count):
        partyinpower = 'Others'
        partyinpower_count = year_max
    
    if (partyinpower == 'Democrats'):
        partyinpower_count = dem_count
    elif (partyinpower == 'Republicans'):
        partyinpower_count = rep_count
    elif (partyinpower == 'Federalists'):
        partyinpower_count = fed_count
    elif (partyinpower == 'Whigs'):
        partyinpower_count = whig_count
    elif (partyinpower == 'Democratic Republicans'):
        partyinpower_count = demrep_count
    elif (partyinpower == 'Independents'):
        partyinpower_count = ind_count
    elif (partyinpower == 'Others'):
        partyinpower_count = oth_count
    
    
    year_parties_pres.append({'year':year, 'democrats':dem_count, 'republicans':rep_count, \
                         'federalist':fed_count, 'whig':whig_count, 'democratic republicans':demrep_count, \
                         'others':oth_count, 'total_members':members_count, 'partyinpower':partyinpower, \
                         'partycount':partyinpower_count, 'independents':ind_count})
 
    year += 1
```

```python
df_years_pres = pd.DataFrame(year_parties_pres)
df_years_pres
```

```python
df_years_pres.to_csv('Presidents.csv', sep=',', encoding='utf-8')
```
