```python
import pandas as pd
import yaml

with open('legislators-merged.yaml', 'r') as f:
    df_terms = pd.io.json.json_normalize(yaml.load(f), 'terms')

df_terms
```

```python
df_senators = df_terms[df_terms.type == 'sen']
df_representatives = df_terms[df_terms.type == 'rep']
```

```python
year = 1789
year_parties_sen = []

while year < 2018:
    
    dem_count = 0
    rep_count = 0
    fed_count = 0
    whig_count = 0
    demrep_count = 0
    ind_count = 0
    oth_count = 0
    members_count = 0
    partyinpower = ""
    partyinpower_count = 0
    
    for index, row in df_senators.iterrows():
        party = row['party']
        year_start = row['start'][:5]
        if (year_start.find('-') > 0):
            year_start = year_start.split("-",1)[0]
        year_end = row['end'][:5]
        if (year_end.find('-') > 0):
            year_end = year_end.split("-",1)[0]
        r = range(int(year_start),int(year_end))
        if (int(year_start) <= year <= int(year_end)):
            if (party == "Democrat"):
                dem_count += 1
            elif (party == "Republican"):
                rep_count += 1
            elif (party == "Federalist"):
                fed_count += 1
            elif (party == "Whig"):
                whig_count += 1
            elif (party == "Democratic Republican"):
                demrep_count += 1
            elif (party == "Independent"):
                ind_count += 1
            else:
                oth_count += 1

    members_count = dem_count + rep_count + fed_count + whig_count + demrep_count + ind_count + oth_count            
                
    year_count_list = [dem_count, rep_count, fed_count, whig_count, demrep_count, ind_count, oth_count]
    year_max = max(year_count_list)
    
    if (year_max == dem_count):
        partyinpower = 'Democrats'
        partyinpower_count = year_max
    
    if (year_max == rep_count):
        partyinpower = 'Republicans'
        partyinpower_count = year_max
    
    if (year_max == fed_count):
        partyinpower = 'Federalists'
        partyinpower_count = year_max
        
    if (year_max == whig_count):
        partyinpower = 'Whigs'
        partyinpower_count = year_max
        
    if (year_max == demrep_count):
        partyinpower = 'Democratic Republicans'
        partyinpower_count = year_max
        
    if (year_max == ind_count):
        partyinpower = 'Independents'
        partyinpower_count = year_max
        
    if (year_max == oth_count):
        partyinpower = 'Others'
        partyinpower_count = year_max
                
    year_parties_sen.append({'year':year, 'democrats':dem_count, 'republicans':rep_count, \
                         'federalist':fed_count, 'whig':whig_count, 'democratic-republicans':demrep_count, \
                         'others':oth_count, 'total_members':members_count, 'partyinpower':partyinpower, \
                         'partycount':partyinpower_count, 'independents':ind_count})
 
    year += 1
```

```python
df_years_sen = pd.DataFrame(year_parties_sen)
df_years_sen
```

```python
df_years_sen.to_csv('Senators.csv', sep=',', encoding='utf-8')
```
