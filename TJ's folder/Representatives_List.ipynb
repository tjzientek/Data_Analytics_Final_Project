```python
import pandas as pd
import json
import os

filepath = os.path.join("representatives.json")
with open(filepath) as jsonfile:
    data_json = json.load(jsonfile)
```

```python
data = data_json
```

```python
datarow = {}
data_list = []
x = 0

while x < len(data[0]):
    first_name = data[0][x]['first_name']
    last_name = data[0][x]['last_name']
    middle_name = data[0][x]['middle_name']
    party = data[0][x]['terms'][0]['party']
    state = data[0][x]['terms'][0]['state']
    year_start = data[0][x]['terms'][0]['year_start']
    year_end = data[0][x]['terms'][0]['year_end']
    datarow = {'first_name':first_name, 'last_name':last_name, 'middle_name':middle_name, 'party':party, \
               'state':state, 'year_start':year_start, 'year_end':year_end}
    data_list.append(datarow)
    x += 1
 
df = pd.DataFrame(data_list)
df.head()
```

```python
df.to_csv("representatives.csv", index=False, header=True)
```

```python
year = 1789
year_parties = []


while year < 2018:
    
    dem_count = 0
    rep_count = 0
    fed_count = 0
    whig_count = 0
    demrep_count = 0
    oth_count = 0
    x = 0
    while x < len(data[0]):
        party = data[0][x]['terms'][0]['party']
        year_start = data[0][x]['terms'][0]['year_start']
        if (year_start.find('-') > 0):
            year_start = year_start.split("-",1)[0]
        year_end = data[0][x]['terms'][0]['year_end']
        if (year_end.find('-') > 0):
            year_end = year_end.split("-",1)[0]
        #print(f"Party: {party}, Year: {year_start}, {year_end}")
        if (party == "Democratic" and int(year_start) <= year <= int(year_end)):
            dem_count += 1
        elif (party == "Republican" and int(year_start) <= year <= int(year_end)):
            rep_count += 1
        elif (party == "Federalist" and int(year_start) <= year <= int(year_end)):
            fed_count += 1
        elif (party == "Whig" and int(year_start) <= year <= int(year_end)):
            whig_count += 1
        elif (party == "Democratic-Republican" and int(year_start) <= year <= int(year_end)):
            demrep_count += 1
        elif (int(year_start) <= year <= int(year_end)):
            oth_count += 1
        x += 1   
    year_parties.append({'year':year, 'democrats':dem_count, 'republicans':rep_count, \
                         'federalist':fed_count, 'whig':whig_count, 'democratic-republicans':demrep_count, \
                         'others':oth_count})
 
    year += 1
```

```python
df_years = pd.DataFrame(year_parties)
df_years
```
