```python
import pandas as pd
```

```python
dfHOR = pd.read_csv('HOR.csv')
dfHOR.head()
```

```python
dfTreasury = pd.read_csv('AC_30-year-treasury-bond-rate-yield-chart.csv')
dfTreasury.head()
```

```python
dfPresident = pd.read_csv('AC_Presidential_Party.csv')
dfPresident.head()
```

```python
dfAlcohol = pd.read_csv('Alcohol_Consumption.csv')
dfAlcohol.head()
```

```python
dfSenate = pd.read_csv('senate.csv')
dfSenate.head()
```

```python
dfBread = pd.read_csv('Bread_prices.csv')
dfBread.head()
```

```python
Comb1 = pd.merge(left=dfSenate, right=dfHOR, how='outer')
Comb1.head()
```

```python
Comb2= pd.merge(left=dfTreasury, right=dfPresident, how='outer')
Comb2.head()
```

```python
Comb3= pd.merge(left=Comb1, right=Comb2, how='outer')
Comb3.head()
```

```python
Comb4 = pd.merge(left=Comb3, right=dfAlcohol, how='outer')
Comb4.head()
```

```python
Combine_csv= pd.merge(left=Comb4, right=dfBread, how='outer')
Combine_csv
```

```python
Combine_csv.to_csv('Combine_csv.csv', sep = ',')
```
